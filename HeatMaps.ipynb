```python
#!pip install seaborn==0.12.2
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive
heatmap_data = data2[['Term','sad_count','happy_count','Neutral_count','Frequency']].dropna(subset = ['Term'])
heatmap_data = heatmap_data.set_index('Term')
df = heatmap_data.copy()
idx = ['sad_count', 'happy_count', 'Neutral_count','Frequency']
df[idx] = df[idx].apply(lambda x: x * 100 / x['Frequency'], axis=1)
df = df.drop('Frequency', axis=1)


#Plot a heatmap
#sns.heatmap(df, cmap="YlGnBu", annot=True, fmt=".1f", linewidths=.5,
#            cbar_kws={'label': 'Percentage'})
sns.heatmap(df, cmap="YlGnBu", annot=True, fmt=".1f", linewidths=.5,
            cbar_kws={'label': 'Percentage', 'shrink': 0.8}, # Adjusted colorbar
            annot_kws={"size": 12}) # Increased annotation font size


# Plot the stacked bar graph
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
df.plot(kind='bar', stacked=True)
plt.xlabel('Term')
plt.ylabel('Percentage')
plt.title('Stacked Bar Graph of Sentiment Counts')
plt.xticks(rotation=90)
plt.legend(loc='upper left', bbox_to_anchor=(1.0, 1.0))
plt.tight_layout()
plt.show()


```

```python
#!pip install seaborn==0.12.2
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from google.colab import drive

drive.mount('/content/gdrive')
data2 = pd.read_excel(r'/content/gdrive/My Drive/ThuInternship/terms.xlsx', sheet_name='2022_2024')

# Assuming keywords are in column A and 'sad_count' is in another column
keywords = data2.iloc[:, 0]  # Get values from column A (index 0)
sad_counts = data2['sad_count']  # Replace 'sad_count' with the actual column name

# Create a DataFrame for the heatmap
heatmap_data = pd.DataFrame({'keyword': keywords, 'sad_count': sad_counts})
heatmap_data = heatmap_data.set_index('keyword')

plt.figure(figsize=(10, 8))
sns.heatmap(heatmap_data, annot=True, cmap='viridis', linewidths=.5)
plt.title('Heatmap of Sad Count by Keyword')
plt.show()
```
